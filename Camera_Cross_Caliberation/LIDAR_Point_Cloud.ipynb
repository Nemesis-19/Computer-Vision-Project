```python
# import the libraries
import numpy as np
import cv2 as cv
import glob
import matplotlib.pyplot as plt
import open3d as o3d
from google.colab import drive
drive.mount('/content/drive')
```

```python
# collecting LIDAR images using glob library
images = glob.glob('lidar_data/*.pcd')

for fname in images:
    print(fname)
    # creating empty PLY Point Cloud (pcd) object
    ply_point_cloud = o3d.data.PLYPointCloud()
    # storing the pcd image on this point
    pcd = o3d.io.read_point_cloud(fname)
    print(pcd, "\n")
    # calculating normal vector usiong svd
    U, s, Vd = np.linalg.svd(np.asarray(pcd.points) - np.mean(np.asarray(pcd.points), axis=0))
    print("\nNormal Vector:", Vd[-1])
    # calculating offset by taking dot product of normal vector and centroid of point cloud
    offset = np.dot(Vd[-1], np.mean(np.asarray(pcd.points), axis=0))
    print("Offset:", offset)
    print("\n-------------------------------\n")
```
